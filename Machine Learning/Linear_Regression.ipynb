```python
#Sample heights in centimeters 
X = [[151],[174],[138],[186],[128],[136],[179],[163],[152],[131]]
#Corresponding weights in Kilograms
y = [63,81,56,91,47,57,76,72,62,48]
```

```python
#Import Linear Regression Class from scikit-Learn
from sklearn.linear_model import LinearRegression
#Create Linear Regression object
regressor = LinearRegression()
#Train the algorithm with given data (i/p and o/p)
regressor.fit(X,y)
```

```python
new = [[175]]
pred = regressor.predict(new)
print('Weight of the person will be',pred[0],'kgs')
```

```python
print("Accuracy:",regressor.score(X,y)*100)
```

```python
print(regressor.coef_)
print(regressor.intercept_)
```
